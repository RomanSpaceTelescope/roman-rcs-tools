# WFI RCS Analysis Notebooks

## Overview
This collection contains analysis notebooks for the Wide Field Instrument Relative Calibration System characterization and performance validation.

---

## 📚 Notebook Directory

#### [AFT.ipynb](AFT.ipynb)
**Purpose**: Ambient Functional Test (AFT) long-term trending and LED health monitoring  
**Key Output**: Time-series plots of LED/photodiode performance, settling metrics, heater validation  
**Use When**: Tracking LED degradation, pre/post-TVAC comparisons, routine health checks  
**Run Time**: ~1-2 min (read existing) / ~5 min (reprocess)  
**Dependencies**: AFT telemetry data in `data/aft/aft_telem.h5`

#### [LED Chargeup.ipynb](LED%20Chargeup.ipynb)
**Purpose**: Analyze LED charge-up process and settling improvements  
**Key Output**: 60to300 settling metrics, before/after comparison plots  
**Use When**: Evaluating LED conditioning protocols or investigating drift  
**Run Time**: ~20 seconds (load existing) / ~5 min (reprocess)

#### [LED SpectraPlot.ipynb](LED_SpectraPlot.ipynb)
**Purpose**: Publication quality spectrum plot
**Key Output**: SCA, PD, LED quantum efficiency and spectrum
**Use When**: Making publication plots of LED spectrum for distribution
**Run Time**: ~20 seconds

#### [PreChargesRCS.ipynb](PreChargesRCS.ipynb)
**Purpose**: Calculate optimal precharge parameters and map effectiveness  
**Key Output**: Precharge calculator, settling vs precharge plots  
**Use When**: Planning observation sequences or optimizing LED protocols  
**Run Time**: ~20 seconds (load existing) / ~5 min (reprocess)

#### [RCS_Telemetry_from_bins.ipynb](RCS_Telemetry_from_bins.ipynb)
**Purpose**: Document the process of extractings RCS telemetry from raw SDR bins
**Key Output**: Markdown documentation
**Use When**: Trying to understand how to process data
**Run Time**: N/A - just documentation of steps

#### [SCIPA_TVAC_Tuning.ipynb](SCIPA_TVAC_Tuning.ipynb)
**Purpose**: RCS tuning analysis comparing TVAC1, TVAC2, and SCIPA campaigns  
**Key Output**: Code→flux calibration curves, cross-campaign bar charts  
**Use When**: Validating RCS calibration or investigating performance changes  
**Run Time**: ~5 min  
**Dependencies**: Requires TVAC data files in `data/tuning/`

#### [sRCS_PersonalityTable.ipynb](sRCS_PersonalityTable.ipynb)
**Purpose**: Browse LED personality tables, which show LED by LED performance and settings
**Key Output**: Command tables with code→flux mappings  
**Use When**: Updating flight software or validating commanded fluxes
